```python
import cv2
import numpy as np
import gc

# Load video
video_path = '/kaggle/input/input-video/WhatsApp Video 2025-04-05 at 06.05.36_dbb7f852.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Cannot open video")
    exit()

print("Video opened successfully")

# Resize settings
resize_width, resize_height = 640, 360

# Background subtractor for motion detection
fgbg = cv2.createBackgroundSubtractorMOG2(history=200, varThreshold=25, detectShadows=True)

# Heatmap accumulation
heatmap_accum = np.zeros((resize_height, resize_width), dtype=np.float32)

# Get FPS and total frames
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = 0
max_frames = 300  # Set upper limit to avoid crash in Kaggle

# Define video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out_path = '/kaggle/working/output_processed.mp4'
out = cv2.VideoWriter(out_path, fourcc, fps, (resize_width, resize_height))

while cap.isOpened() and frame_count < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    # Resize
    frame = cv2.resize(frame, (resize_width, resize_height))

    # Motion mask
    fgmask = fgbg.apply(frame)

    # Update heatmap
    heatmap_accum += (fgmask > 0).astype(np.float32)

    # Normalize for visualization
    normalized_heatmap = cv2.normalize(heatmap_accum, None, 0, 255, cv2.NORM_MINMAX)
    colored_heatmap = cv2.applyColorMap(normalized_heatmap.astype(np.uint8), cv2.COLORMAP_JET)

    # Overlay heatmap on original frame
    overlay = cv2.addWeighted(frame, 0.7, colored_heatmap, 0.3, 0)

    # Write to output video
    out.write(overlay)

    # Memory cleanup
    del frame, fgmask, normalized_heatmap, colored_heatmap, overlay
    gc.collect()

    frame_count += 1

# Cleanup
cap.release()
out.release()
gc.collect()

print(f"✅ Processing done! Output saved to: {out_path}")
```
