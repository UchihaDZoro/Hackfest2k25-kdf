```python
import cv2
from ultralytics import YOLO
import os

# Load YOLOv8 model
model = YOLO('yolov8n.pt')  # You can use yolov8s.pt or yolov8m.pt for better accuracy

# Load your video
video_path = '/kaggle/input/d/silentwisper/ytvideo/youtube_TrAUA1xS1Oc_640x480_h264.mp4'  # Change to your uploaded video path
cap = cv2.VideoCapture(video_path)

# Output video config
out_path = 'thermal_detection_output.avi'
fourcc = cv2.VideoWriter_fourcc(*'XVID')
fps = int(cap.get(cv2.CAP_PROP_FPS))
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(out_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO object detection
    results = model(frame, verbose=False)[0]

    # Draw bounding boxes and labels
    for box in results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cls = int(box.cls[0])
        label = model.names[cls]
        confidence = box.conf[0].item()
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f"{label} {confidence:.2f}", (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

    # Convert frame to "thermal" effect (colormap)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    thermal = cv2.applyColorMap(gray, cv2.COLORMAP_JET)

    # Save to output video
    out.write(thermal)

cap.release()
out.release()
print(f"✅ Processed video saved as {out_path}")
```
